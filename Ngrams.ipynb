# Ngrams:
N-grams, sequential patterns of n items (characters or words), aid in contextual analysis, language modeling, and efficient text indexing. They recognize patterns, making them vital for sentiment analysis, text classification, and more in NLP.

```python
import re
```

```python
def Ngrams(text, n):

  tokens=re.split("\\s+", text)

  n_grams=[]
  for i in range(len(tokens)-n+1):
    temp=[]
    for j in range(i,i+n):
      temp.append(tokens[j])
    n_grams.append(" ".join(temp))

  return n_grams
```

```python
sentence ="The quick brown fox jumps over the lazy dog"
```

## Unigrams (1-grams):
These are single items, typically words, in a sequence. Each word in a sentence or document is considered a unigram.

```python
unigrams = Ngrams(sentence, 1)
print(unigrams)
```

## Bigrams (2-grams):
These consist of pairs of consecutive items. In the context of text, bigrams are often used to analyze word pairs that appear together.

```python
bigrams=Ngrams(sentence, 2)

print(bigrams)
```

## Trigrams (3-grams):
These are sequences of three consecutive items. Trigrams can capture more complex relationships between words or characters.

```python
trigrams=Ngrams(sentence, 3)

print(trigrams)
```
