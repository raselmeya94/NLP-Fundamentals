sentence are tokenize using by NLTK

```python
import nltk
nltk.download("punkt")
from nltk.tokenize import word_tokenize
```

```python
sentence="Bangladesh is small country but it\'s population huge and exponentially increasing rather than U.S.A"
```

```python
tokens=word_tokenize(sentence)
print(tokens)
```

NLP Based tokenize using by Spacy

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(sentence)
```

```python
for token in doc:
  print(token, end=" || ")
```

```python
doc4=nlp(u" Bangladesh is small country, its propulation growth is 1.08% increase from 2021 and per capita is $2,687.000 ")
```

```python
for token in doc4:
  print(token, end=" | ")

print("\n===============")

# Named Entity Recognition (NER): Printing the recognized entities along with their labels and explanations

for token in doc4.ents:
  print(token.text+ "    "+ token.label_ + "   "+ str(spacy.explain(token.label_)))
```

The provided code demonstrates tokenization and named entity recognition (NER) using the spaCy library

```python

# Print the noun chunks
for chunk in doc4.noun_chunks:
    print(chunk.text)
```
