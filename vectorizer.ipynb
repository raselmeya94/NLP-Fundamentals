```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# documents

doc_A="The car is driven on the road"
doc_B="The truck is driven on the highway"
```

```python
corpus=[doc_A,doc_B]

print(corpus)
```

```python

#create object of CountVectorizer
vectorizer=CountVectorizer()

vectorizer.fit(corpus)
```

```python
vectorizer.vocabulary_
```

```python
output_vector=vectorizer.transform(corpus)
output=output_vector.toarray()
```

```python
output
```

```python

import pandas as pd
document_names = ['Doc_A', 'Doc_B']
vocabulary_words = vectorizer.get_feature_names_out()

# Create a DataFrame
df = pd.DataFrame(data=output, columns=vocabulary_words, index=document_names)

print(df)
```
