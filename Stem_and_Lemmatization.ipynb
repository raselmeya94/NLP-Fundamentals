# Stemming

1. What is Stemming?

Stemming is a natural language processing technique that **reduces words to their base or root form**, known as the stem.
The stem is obtained by removing suffixes or prefixes from words to achieve a common base form.
2. Purpose of Stemming:

Text Normalization: Stemming helps in normalizing text by reducing inflected or derived words to a common base.
Dimensionality Reduction: By reducing words to their stems, stemming reduces the number of unique word forms in the text, simplifying analysis.
Speed and Efficiency: Stemming is computationally faster compared to lemmatization, making it suitable for large-scale text processing.
3. How Stemming Works:

The stemming process involves applying a set of language-specific rules to remove suffixes and prefixes from words.
The resulting stem may not always be a valid word, but it represents the base form for related words.
4. Examples of Stemming:

Words like "running," "runs," "ran" are stemmed to "run."
Words like "jumps," "jumped," "jumping" are stemmed to "jump."

5. Stemming Libraries:

NLTK: Use the Porter or Snowball stemmers provided by the NLTK library.
spaCy: Although spaCy primarily focuses on lemmatization, you can still use custom stemming techniques if required.
TextBlob: TextBlob provides a basic stemmer for English text.
6. Considerations:

Overstemming: Stemming may result in overly aggressive reductions, causing multiple related words to be stemmed to the same base, leading to loss of context.
Lemmatization vs. Stemming: Lemmatization provides more meaningful word forms compared to stemming, but it is computationally more expensive.
Stemming is a valuable text processing technique for text normalization and dimensionality reduction. While it is less accurate than lemmatization, it remains a popular choice in scenarios where speed and simplicity are critical, such as information retrieval, search engines, and some information extraction tasks. However, it's important to consider the trade-offs and choose the appropriate text processing method based on the specific requirements of your NLP application.

```python
import nltk
```

```python
sentence="Eating eat ate history historical read reading adjustable adjusting"
```

```python
words=sentence.split(" ")
```

```python
print(words)
```

Import Stemmer


```python
from nltk.stem import PorterStemmer
```

```python
# create a stemmer object
stemmer = PorterStemmer()
```

```python
for token in words:
  print(token, " | ", stemmer.stem(token)) # word are converted by the stemmer
```

```python
# Here we can see , (Eating, Ete)-> eat  || reading , read -> read || adjustable, adjusting -> adjust and so on
```

# Lemmatization



*1. What is Lemmatization?*

Lemmatization is a natural language processing technique that reduces words to their base or root form, known as the lemma.
The lemma represents the canonical form of a word, which could be a noun, verb, adjective, or adverb.
*2. Why Lemmatization?*

**Text Normalization:** Lemmatization helps in normalizing text by reducing variations of the same word to a common form.
**Improved Text Analysis:** Lemmatization ensures that different forms of a word are treated as the same word during analysis, leading to more accurate results.
**Information Retrieval:** In information retrieval systems, lemmatization aids in retrieving relevant documents by matching lemmatized query terms to indexed documents.
**Semantic Analysis**: By reducing words to their lemmas, lemmatization captures the semantic meaning of the text.

*3. How to Use Lemmatization?*

**NLTK:** Use the WordNetLemmatizer from NLTK to lemmatize words. Specify the part-of-speech (POS) tag if needed.
**spaCy:** Use the built-in lemmatization capabilities of spaCy for simple and efficient lemmatization.
**TextBlob:** Employ TextBlob for lemmatizing words, considering the POS if required.
**Standalone WordNet:** Utilize WordNetLemmatizer from NLTK with the WordNet corpus for lemmatization and POS tagging.



Lemmatization is a crucial step in text preprocessing that aids in text normalization and enhances various NLP tasks like text classification, sentiment analysis, and information retrieval. By reducing words to their base forms, lemmatization simplifies text analysis and improves language understanding in NLP applications.

```python
from nltk.stem import  WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('punkt')
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
for token in words:
  print(token, " | ", lemmatizer.lemmatize(token))
```
